```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(sparse=False, drop='first')
item_category_encoded = one_hot_encoder.fit_transform(final[['item_category_id']])

feature_names = one_hot_encoder.get_feature_names_out(['item_category_id'])
one_hot_encoded_df = pd.DataFrame(item_category_encoded, columns=feature_names)

final = pd.concat([final, one_hot_encoded_df], axis=1)
```

```python
final['date'] = pd.to_datetime(final['date'], format='%d.%m.%Y')
final['month'] = final['date'].dt.month
final['day_of_week'] = final['date'].dt.dayofweek
final['item_cnt_day'] = final['item_cnt_day'].apply(lambda x: max(x, 0))
```

```python
X = final[['date_block_num', 'shop_id', 'item_id', 'item_price', 'month', 'day_of_week']]
y_revenue = final['item_price'] * final['item_cnt_day']  # Assuming item_price * item_cnt_day gives revenue
y_volume = final['item_cnt_day']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_revenue_train, y_revenue_test, y_volume_train, y_volume_test = train_test_split( X, y_revenue, y_volume, test_size=0.2, random_state=42)
```

```python
model_revenue = RandomForestRegressor(random_state=42)
model_volume = RandomForestRegressor(random_state=42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
y_imputer = SimpleImputer(strategy='mean')
y_revenue_train_imputed = y_imputer.fit_transform(y_revenue_train.values.reshape(-1, 1)).flatten()
y_revenue_test_imputed = y_imputer.transform(y_revenue_test.values.reshape(-1, 1)).flatten()

```

```python
model_revenue = RandomForestRegressor(random_state=42)
model_revenue.fit(X_train_imputed, y_revenue_train_imputed)

```

```python
y_revenue_pred = model_revenue.predict(X_test_imputed)

mse_revenue = mean_squared_error(y_revenue_test_imputed, y_revenue_pred)
print(f'Mean Squared Error (Revenue): {mse_revenue}')
```

```python
import numpy as np

rmse_revenue = np.sqrt(mse_revenue)
print(f'Root Mean Squared Error (Revenue): {rmse_revenue}')
```

```python
y_revenue_pred_linear = model_linear.predict(X_test_imputed)
y_volume_pred_linear = model_linear.predict(X_test_imputed)

mse_revenue_linear = mean_squared_error(y_revenue_test_imputed, y_revenue_pred_linear)
mse_volume_linear = mean_squared_error(y_volume_test, y_volume_pred_linear)

rmse_revenue_linear = np.sqrt(mse_revenue_linear)

print(f'Mean Squared Error (Revenue) - Linear Regression: {mse_revenue_linear}')
print(f'Mean Squared Error (Volume) - Linear Regression: {mse_volume_linear}')
print(f'Root Mean Squared Error (Revenue) - Linear Regression: {rmse_revenue_linear}')
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

```

```python
model_xgboost = XGBRegressor(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_child_weight': [1, 2, 3]
}

```

```python
grid_search = GridSearchCV(estimator=model_xgboost, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)

grid_search.fit(X_train_imputed, y_revenue_train_imputed)

```

```python
best_params = grid_search.best_params_

best_model_xgboost = grid_search.best_estimator_

y_revenue_pred_best = best_model_xgboost.predict(X_test_imputed)

```

```python
mse_revenue_best = mean_squared_error(y_revenue_test_imputed, y_revenue_pred_best)
rmse_revenue_best = np.sqrt(mse_revenue_best)

print(f'Best Hyperparameters: {best_params}')
print(f'Mean Squared Error (Revenue) - Best Model: {mse_revenue_best}')
print(f'Root Mean Squared Error (Revenue) - Best Model: {rmse_revenue_best}')
```
